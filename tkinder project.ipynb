```python
from tkinter import *
import pymysql
import re


def clear_fields():
    e1.delete(0, END)
    e2.delete(0, END)

def sign_up():
    email = e1.get()
    password = e2.get()
    
    if validate_email(email) and validate_password(password):
        nam(connection, email, password)
        error_label.config(text="Sign-up successful.")
        clear_fields()
    else:
        error_label.config(text="Invalid email or password format. Please check.")

def is_user_registered(connection, email, password):
    cursor = connection.cursor()
    cursor.execute("SELECT * FROM user WHERE email = %s AND passcode = %s", (email, password))
    return cursor.fetchone() is not None

def validate_email(email):
    email_pattern = r'^[a-zA-Z0-9._%+-]+@[a-zAZ0-9.-]+\.[a-zA-Z]{2,}$'
    return re.match(email_pattern, email) is not None

def validate_password(password):
    return len(password) >= 6

def nam(connection, email, password):
    cursor = connection.cursor()
    cursor.execute("INSERT INTO user (email, passcode) VALUES (%s, %s)", (email, password))
    connection.commit()
    print("User Record inserted successfully.")

def insert_data(connection, data):
    cursor = connection.cursor()
    cursor.execute("INSERT INTO form (first_name, last_name, email, dob, gender, address, city, state, pincode, phone_number, degree, branch, year_of_passing, skills) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)", data)
    connection.commit()
    print("Form Data inserted successfully.")

def open_sign_up_window():
    signup_window = Toplevel(r)
    signup_window.title("Sign Up")
    
    # Add labels and entry fields for additional information
    first_name_label = Label(signup_window, text='First Name', width=20)
    last_name_label = Label(signup_window, text='Last Name', width=20)
    email_label = Label(signup_window, text='Email', width=20)
    dob_label = Label(signup_window, text='DoB', width=20)
    gender_label = Label(signup_window, text='Gender', width=20)
    address_label = Label(signup_window, text='Address', width=20)
    city_label = Label(signup_window, text='City', width=20)
    state_label = Label(signup_window, text='State', width=20)
    pincode_label = Label(signup_window, text='Pincode', width=20)
    phone_number_label = Label(signup_window, text='Phone Number', width=20)
    degree_label = Label(signup_window, text='Degree', width=20)
    branch_label = Label(signup_window, text='Branch', width=20)
    year_of_passing_label = Label(signup_window, text='Year of Passing', width=20)
    skill_label = Label(signup_window, text='Skills', width=20)
    
    first_name_label.grid(row=0, column=0)
    last_name_label.grid(row=1, column=0)
    email_label.grid(row=2, column=0)
    dob_label.grid(row=3, column=0)
    gender_label.grid(row=4, column=0)
    address_label.grid(row=5, column=0)
    city_label.grid(row=6, column=0)
    state_label.grid(row=7, column=0)
    pincode_label.grid(row=8, column=0)
    phone_number_label.grid(row=9, column=0)
    degree_label.grid(row=10, column=0)
    branch_label.grid(row=11, column=0)
    year_of_passing_label.grid(row=12, column=0)
    skill_label.grid(row=13, column=0)
    
    global e3, e4, e5, e6, e7, e8, e9, e10, e11, e12, e13, e14, e15, e16
    
    e3 = Entry(signup_window, width=30)
    e4 = Entry(signup_window, width=30)
    e5 = Entry(signup_window, width=30)
    e6 = Entry(signup_window, width=30)
    e7 = Entry(signup_window, width=30)
    e8 = Entry(signup_window, width=30)
    e9 = Entry(signup_window, width=30)
    e10 = Entry(signup_window, width=30)
    e11 = Entry(signup_window, width=30)
    e12 = Entry(signup_window, width=30)
    e13 = Entry(signup_window, width=30)
    e14 = Entry(signup_window, width=30)
    e15 = Entry(signup_window, width=30)
    e16 = Entry(signup_window, width=30)
    
    e3.grid(row=0, column=1)
    e4.grid(row=1, column=1)
    e5.grid(row=2, column=1)
    e6.grid(row=3, column=1)
    e7.grid(row=4, column=1)
    e8.grid(row=5, column=1)
    e9.grid(row=6, column=1)
    e10.grid(row=7, column=1)
    e11.grid(row=8, column=1)
    e12.grid(row=9, column=1)
    e13.grid(row=10, column=1)
    e14.grid(row=11, column=1)
    e15.grid(row=12, column=1)
    e16.grid(row=13, column=1)
    
    button_signup = Button(signup_window, text="Submit", bg='Blue', fg='White', command=submit_sign_up)
    button_signup.grid(row=14, columnspan=2)
    
    button_cancel = Button(signup_window, text="Cancel", bg='red', fg='white', command=signup_window.destroy)
    button_cancel.grid(row=14, column=2)

def submit_sign_up():
    # Retrieve and store the user information
    first_name = e3.get()
    last_name = e4.get()
    email = e5.get()
    dob =e6.get()
    gender = e7.get()
    address = e8.get()
    city = e9.get()
    state = e10.get()
    pincode = e11.get()
    phone_number = e12.get()
    degree = e13.get()
    branch = e14.get()
    year_of_passing = e15.get()
    skill = e16.get()
    
    data = (first_name, last_name, email, dob, gender, address, city, state, pincode, phone_number, degree, branch, year_of_passing, skill)
    
    insert_data(connection, data)
    
    # Open a new window to display a success message
    show_success_message()

def show_success_message():
    success_window = Toplevel(r)
    success_window.title("Success")
    success_label = Label(success_window, text=" successful Inserted.")
    success_label.config(bg='lightgreen') 
    success_label.pack()
    clear_fields()

# Database connection
connection = pymysql.connect(host='localhost', user='root', password='Krishna@2002', db='Project')

# Main window
r = Tk()
r.title("krish&Co")
x = Label(r, text='Email')
y = Label(r, text='Password')
x.grid(row=0)
y.grid(row=1)
e1 = Entry(r, width=30)
e2 = Entry(r, width=30, show='*')
e1.grid(row=0, column=1)
e2.grid(row=1, column=1)
button_signup = Button(r, text="Sign Up", bg='Blue', fg='White', command=sign_up)
button_signup.grid(row=2, column=0)
button_signin = Button(r, text="Sign In", bg='green', fg='white', command=open_sign_up_window)
button_signin.grid(row=2, column=1)
button_cancel = Button(r, text="Cancel", bg='red', fg='white', command=clear_fields)
button_cancel.grid(row=2, column=2)

# Label to display error messages
error_label = Label(r, text="", fg="red")
error_label.grid(row=4, columnspan=2)

# Main event loop
r.mainloop()

# Close the database connection
connection.close()
```
